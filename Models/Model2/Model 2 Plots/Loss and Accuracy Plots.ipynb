```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

# Plots train and validation loss and accuracy for model 1 (model already trained)
# A
df_A = pd.read_csv("A.out", delimiter="\n", engine='python', header= None)
df_A = df_A.dropna(axis=1)

df_A = df_A[df_A[0] != "----------"]

df_A = df_A[df_A[0]!= "model5_test4"]

train_dict = {}

number = 0
for i in df_A[0]:
    if "train" in i:
        train_dict[number] = i.split(" ")
        number += 1

train_acc_dict = {}
for i, j in train_dict.items():
    train_acc_dict[i]=float(j[-1])

train_loss_dict = {}
for i, j in train_dict.items():
    train_loss_dict[i]=float(j[-3])

val_dict = {}

number = 0
for i in df_A[0]:
    if "validate" in i:
        val_dict[number] = i.split(" ")
        number += 1

val_acc_dict = {}
for i, j in val_dict.items():
    val_acc_dict[i]=float(j[-1])

val_loss_dict = {}
for i, j in val_dict.items():
    val_loss_dict[i]=float(j[-3])

# plot accuract
fig, ax = plt.subplots()
ax.plot(train_acc_dict.values(),label='train accuracy')
ax.plot(val_acc_dict.values(),label='validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title("Agreeableness Prediction: Accuracy")
ax.legend()
plt.show() 

# plot loss
fig, ax = plt.subplots()
ax.plot(train_loss_dict.values(),label='train loss')
ax.plot(val_loss_dict.values(),label='validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title("Agreeableness Prediction: Loss")
ax.legend()
plt.show() 

```
